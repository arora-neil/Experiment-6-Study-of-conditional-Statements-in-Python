```python
# 2.Write a python program to check if given number is either even or odd.
a=int(input("Enter a Number: "))
if a%2==0:
  print("Number is Even")
else:
  print("Number is Odd")
```

```python
# 3. To find the largest number out of the given numbers
a=int(input("Enter a: "))
b=int(input("Enter b: "))
c=int(input("Enter c: "))
if(a>b and a>c):
  print("a is largest")
elif(b>a and b>c):
  print("b is largest")
elif(c>a and c>b):
  print("c is largest")
else:
  print("a,b,c are equal")
```

```python
# 4. Get Input from user for 1 subject and calculate the grade using if-elif-else
m=float(input("Enter marks: "))
if m>=90:
  print("Grade: A")
elif m>=75:
  print("Grade: B")
elif m>=60:
  print("Grade: C")
elif m>=40:
  print("Grade: D")
else:
  print("Fail")
```

```python
# 5. Write python program to check if given year is leap year or not
year=int(input("Enter the year: "))
if( year%4==0 and year%100!=0 or year%400==0):
  print("Leap Year")
else:
 print("Not Leap Year")
```
