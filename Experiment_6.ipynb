```python
'''Problem Statement
Write a program to calculate the gross salary of an employee based
if basic <=10000 : HRA = 20%, DA= 80%
if basic <= 20000 : HRA = 25%, DA = 90%
otherwise: HRA = 30%, DA = 95%
Gross salary=basic+DA+HRA'''
basic=int(input("Enter Basic Salary: "))
if(basic <= 10000):
 HRA,DA=basic*0.20,basic*0.80
elif(basic <= 20000):
  HRA,DA=basic*0.25,basic*0.90
else:
  HRA,DA=basic*0.30,basic*0.95
print("Gross Salary =",basic+HRA+DA)
```
